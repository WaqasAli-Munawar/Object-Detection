Object detection has a close relationship with analysing videos and images, which is why it has gained a lot of attention to so many researchers in recent years. 

To gain a full understanding of the image, we should not only focus on classifying the different images but also try to accurately estimate the concepts and locations of the objects contained in each image. This task is known as **object detection**. Detecting Objects usually consists of different subtasks such as **face detection**, **pedestrian detection**, **Number plate detection** and **skeleton detection**.

As one of the fundamental problems of computer vision, object detection is able to provide valuable information for the semantic understanding of images and videos and is related to many applications, including the **classification of images**, **analysis of human behaviour**, **facial recognition** and **autonomous driving**.

Inheriting from **neural networks** and other machine learning systems, advancements in these areas will allow the development of neural networks and will also have great impacts on the techniques of detecting objects that can be considered as the future machine learning systems.

Now let’s load and prepare the dataset

```python
img_h, img_w, num_channels = (380, 676, 3)
```

```python
df = pd.read_csv('data/train_solution_bounding_boxes (1).csv')
df.rename(columns={'image':'image_id'}, inplace=True)
```

```python
df['image_id'] = df['image_id'].apply(lambda x: x.split('.')[0])
df['x_center'] = (df['xmin'] + df['xmax'])/2
df['y_center'] = (df['ymin'] + df['ymax'])/2
df['w'] = df['xmax'] - df['xmin']
df['h'] = df['ymax'] - df['ymin']
df['classes'] = 0
df['x_center'] = df['x_center']/img_w
df['w'] = df['w']/img_w
df['y_center'] = df['y_center']/img_h
df['h'] = df['h']/img_h
```

Now let’s have a look at the sample image from the dataset before moving further:

```python
index = list(set(df.image_id))
image = random.choice(index)
print("Image ID: %s"%(image))
img = cv2.imread(f'data/training_images/{image}.jpg')
```

```python
Image(filename=f'data/training_images/{image}.jpg',width=600)
```

Now let’s train the machine learning model for detecting objects:

```python
p = %pwd
p
```

```python
source = 'training_images'

if True:
    for fold in [0]:
        val_index = index[len(index)*fold//5:len(index)*(fold+1)//5]
        for name,mini in tqdm(df.groupby('image_id')):
            if name in val_index:
                path2save = 'val2017/'
            else:
                path2save = 'train2017/'
            if not os.path.exists(p+'/convertor/labels/'+path2save):
                os.makedirs(p+'/convertor/labels/'+path2save)
                
            with open(p+'/convertor/labels/'+path2save+name+".txt", 'w+') as f:
                row = mini[['classes','x_center','y_center','w','h']].astype(float).values
                row = row.astype(str)
                for j in range(len(row)):
                    text = ' '.join(row[j])
                    f.write(text)
                    f.write("\n")
                    
            if not os.path.exists(p+'/convertor/images/{}'.format(path2save)):
                os.makedirs(p+'/convertor/images/{}'.format(path2save))
            sh.copy("data/{}/{}.jpg".format(source,name),p+'/convertor/images/{}/{}.jpg'.format(path2save,name))
```

### Detecting Objects

We have successfully trained our model, now we need to test the model on the images to see if our model is working well to detect objects in the images:

```python
predicted_files = []

for (dirpath, dirnames, filenames) in os.walk(p+"\\data\\testing_images"):
    predicted_files.extend(filenames)
```
